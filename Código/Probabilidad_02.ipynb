**Ejemplo**

En una universidad, el 60% de los estudiantes son mujeres y el 40% son hombres. De las mujeres, el 30% estudian ciencias y de los hombres, el 50% estudian ciencias. Si se selecciona al azar un estudiante de la universidad, ¿cuál es la probabilidad de que sea un hombre que estudia ciencias?

**Solución**

En una universidad, el 60% de los estudiantes son mujeres y el 40% son hombres.

$A = \{\text{Ser mujer}\}$

$B = \{\text{Ser hombre}\}$

$P(A) = 0.60$

$P(B) = 0.40$

De las mujeres, el 30% estudian ciencias y de los hombres, el 50% estudian ciencias.

$C = \{\text{Estudiar ciencias}\}$

$P(C|A) = 0.30$

$P(C|B) = 0.50$

Si se selecciona al azar un estudiante de la universidad, ¿cuál es la probabilidad de que sea un hombre que estudia ciencias?

$P(B \cap C) = ??$

Recondemos que por probabilidad condicional:

$$P(C|B) = \cfrac{P(C \cap B)}{P(B)}$$

Podemos despejar la probabilidad conjunta:

$$P(C \cap B) = P(C|B) \cdot P(B)$$

Por lo tanto

$$P(C \cap B) = 0.50 \cdot 0.40$$

```python
0.50*0.40
```

### En general

$$P(B|A) = \cfrac{P(B\cap A)}{P(A)}$$

y podemos despejar para tener que:

$$P(A\cap B) = P(B|A)\times P(A)$$

**Por tanto, la probabilidad de seleccionar un estudiante al azar de la universidad y que sea un hombre que estudia ciencias es de 0.2.**

# Teorema de Bayes

El teorema de Bayes es un principio fundamental en la teoría de la probabilidad que describe cómo actualizar la probabilidad de un evento basándose en nueva información o evidencia.

El teorema establece que la probabilidad de un evento condicionado a la ocurrencia de otro evento se puede calcular a partir de la probabilidad del evento condicionante y la probabilidad condicional inversa.

La fórmula del teorema de Bayes es la siguiente:

$$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}$$

donde $P(A|B)$ es la probabilidad de que A ocurra dado que B ha ocurrido,$ P(B|A)$ es la probabilidad de que B ocurra dado que A ha ocurrido,  $P(A)$ es la probabilidad inicial de que A ocurra, y $P(B)$ es la probabilidad de que B ocurra.

En resumen, el teorema de Bayes es una herramienta valiosa para calcular la probabilidad condicional de eventos y es ampliamente utilizado en la estadística, la ciencia de datos y otras disciplinas para la toma de decisiones y la inferencia probabilística.

4. La otra probabilidad condicional

**Ejemplo**

Una enfermedad está presente en el 15% de la población. Se tiene una prueba para detectarla que da positivo en el 97% de los enfermos (sensibilidad) y da negativo en el 95% de los sanos (especificidad). Se aplica la prueba a una persona, ¿Cuál es la probabilidad que de positivo?, ¿Cuál es la probabilidad que esté enfermo, dado que la prueba dio positivo?

**Solución**

1. Reconocer eventos


$A = \{\text{Tener la enfermedad}\}$

$B = \{\text{No tener la enfermedad}\}$

$P(A) = 0.15$

$P(B) = 0.85$

2. Reconocer el nuevo evento y las probabilidades que surgen de él.


$C = \{\text{Prueba positiva}\}$

$D = \{\text{Prueba negativa}\}$

Se tiene una prueba para detectarla que da positivo en el 97% de los enfermos (sensibilidad) y da negativo en el 95% de los sanos (especificidad).

$P(C|A) = 0.97$

$P(D|B) = 0.95$



**Pregunta**

Se aplica la prueba a una persona, ¿Cuál es la probabilidad que de positivo?

* La pregunta es $P(C)$

Podemos encontrar dos casos en los que encontraríamos una prueba positiva:

1. Personas que enfermas y que tienen prueba positiva. $P(A \cap C)$
2. Personas que son sanas y que tienen prueba positiva. $P(B \cap C)$


Es decir que usando la probabilidad condicional tenemos:

$$P(C|A) = \cfrac{P(A \cap C)}{P(A)}$$

Despejando:
$$P(A \cap C) = P(C|A) \cdot P(A)$$



En nuestro ejemplo:

$P(A \cap C) =$

```python
0.97*0.15
```

Pero también tenemos que:

$$P(C|B) = \cfrac{P(B \cap C)}{P(B)}$$

Despejando:
$$P(B \cap C) = P(C|B) \cdot P(B)$$

En nuestro ejemplo:

$P(B \cap C) =$

```python
0.05*0.85
```

Quiere decir que en total la probablidad de que una persona tenga prueba positiva, sin importar si está sano o no es de

$$P(C) = P(A\cap C)+P(B\cap C) =$$

```python
0.1455+0.0425
```

**Pregunta**

Se aplica la prueba a una persona, ¿Cuál es la probabilidad que esté enfermo, dado que la prueba dio positivo?

$$P(A|C)$$

Su fórmula condicional es:

$$P(A|C) = \cfrac{P(A\cap C)}{P(C)}$$

Pero como ya sabemos:

$P(A\cap C)= 0.1455$

$P(C) = 0.188$

Por tanto, la probabilidad de que la persona realmente esté enferma dado que la prueba salió positiva es:

```python
0.1455/0.188
```

**Hagamos la tabla en Python**

```python
import pandas as pd
Eventos = ["Enfermo", "Sano"]
Probabilidad = [0.15, 0.85]
Condicionales_anteriores = [0.97, 0.05]
Tabla = pd.DataFrame({
    'Eventos': Eventos,
    'Probabilidad': Probabilidad,
    'Condicionales_anteriores': Condicionales_anteriores
})
Tabla
```

```python
Tabla["Conjuntas"] = Tabla["Probabilidad"]*Tabla['Condicionales_anteriores']
Tabla
```

```python
sum(Tabla['Conjuntas'])
```

```python
Tabla["Posteriores"] = Tabla["Conjuntas"]/sum(Tabla['Conjuntas'])
Tabla
```

**Conclusiones**

La probabilidad de que
1. La persona efectivamente está enferma dado que la prueba salío positiva es de 0.77.
2. La persona no está enferma dado que que la prueba salío positiva es de 0.23.

```python
total_por_columna = Tabla.sum(axis=0)
total_por_columna.name = 'Total'
pd.concat([Tabla, total_por_columna.to_frame().T], axis=0)
```
