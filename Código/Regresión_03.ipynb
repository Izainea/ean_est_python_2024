# Regresión lineal Múltiple

La regresión lineal múltiple permite generar un modelo en el que el valor de la variable dependiente o respuesta $Y$ se determina a partir de un conjunto de variables independientes llamadas predictores $X_1, X_2, X_3, ..., X_p$. Es una extensión de la regresión lineal simple, por lo que es fundamental comprender esta última.

## Módelo

$$Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \cdots + \beta_nX_p + \epsilon$$

Donde:
* $Y$ es la variable dependiente que queremos predecir.
* $X_1, X_2, \ldots, X_n$ son las variables independientes o explicativas.
* $\beta_0$ es el término de intercepción.
* $\beta_1, \beta_2, \ldots, \beta_n$ son los coeficientes que representan el cambio esperado en $Y$ por una unidad de cambio en las variables independientes respectivas.
* $\epsilon$ es el término de error, que captura toda la variabilidad en $Y$ que no puede ser explicada por las variables independientes.

## Supuestos del modelo

Para que las estimaciones de los parámetros del modelo sean no solo eficientes sino también no sesgadas y consistentes, la regresión lineal múltiple asume lo siguiente:

* **Linealidad**: La relación entre las variables independientes y la variable dependiente debe ser lineal. Esto implica que los cambios en las variables independientes tienen efectos aditivos y constantes en la variable dependiente.

* **Independencia de los errores**: Los errores en las predicciones del modelo (los residuos) deben ser independientes entre sí.

* **Homocedasticidad**: La varianza de los errores debe ser constante para todos los niveles de las variables independientes. Si la varianza no es constante (heterocedasticidad), esto puede llevar a estimaciones de los coeficientes que son menos precisas.

* **Normalidad de los errores:** Los errores deben estar aproximadamente normalmente distribuidos, especialmente para el uso de pruebas de hipótesis que se basan en esta suposición para proporcionar estimaciones exactas.

No multicolinealidad: Las variables independientes no deben estar demasiado correlacionadas entre sí. La multicolinealidad puede hacer que los coeficientes de regresión sean inestables y sus interpretaciones difíciles.

### Interpretación de Coeficientes en la Regresión Lineal Múltiple

En la regresión lineal múltiple, cada coeficiente $\beta_i $ muestra el cambio esperado en la variable dependiente $ Y $ por cada unidad de cambio en la variable independiente $ X_i $, manteniendo todas las otras variables independientes constantes. Esta es la clave: "todas las otras variables se mantienen constantes".

**Interpretación de $\beta_i $:**
- **$ \beta_0 $ (Intercepto):** Representa el valor esperado de $ Y $ cuando todas las variables independientes $ X_i $ son cero.
- **$ \beta_i $ (para i = 1, 2, ..., n):** Muestra cómo cambia el valor esperado de $ Y $ con un cambio de una unidad en $ X_i $, asumiendo que todas las otras variables independientes permanecen constantes. Si $ \beta_i $ es positivo, un incremento en $ X_i $ tiende a aumentar $ Y $; si $ \beta_i $ es negativo, un incremento en $ X_i $ tiende a disminuir $ Y $.






### Ejemplo

Supongamos que estamos modelando el salario anual (Y) basado en años de educación ($X_1$) y años de experiencia ($X_2$). La ecuación del modelo podría ser:

$$\text{Salario} = β_0 + β_1(\text{Años de Educación}) + β_2(\text{Años de Experiencia}) + ε$$

$$Y= β_0 + β_1X_1 + β_2X_2 + ε$$


Después de ajustar el modelo con los datos, obtenemos los siguientes coeficientes:

- $ \beta_0 = 20,000 $ (dólares)
- $ \beta_1 = 3,000 $ (dólares por año de educación)
- $ \beta_2 = 2,000 $ (dólares por año de experiencia)

**Interpretación:**
- **Intercepto $ \beta_0 = 20,000 $ dólares:** Si una persona no tiene años de educación ni de experiencia (lo cual es una extrapolación poco realista en este contexto), su salario esperado sería de 20,000 dólares al año.
- **Coeficiente de educación $ \beta_1 = 3,000 $ dólares:** Por cada año adicional de educación, se espera que el salario aumente en 3,000 dólares, asumiendo que los años de experiencia no cambian.
- **Coeficiente de experiencia $ \beta_2 = 2,000 $ dólares:** Por cada año adicional de experiencia laboral, se espera que el salario aumente en 2,000 dólares, manteniendo constante el nivel de educación.



### ¿Cómo se encuentran los coeficientes?

Para encontrar los coeficientes $b_i$ se organizan los datos de la siguiente forma:
1. Matriz con los datos de las variables independientes:
$$X= \begin{bmatrix}
1&x_{11}&x_{12}& \cdots & x_{1p}\\
1&x_{21}&x_{22}& \cdots & x_{2p}\\
\vdots&\vdots&\vdots& \ddots & \vdots\\
1&x_{n1}&x_{n2}& \cdots & x_{np}\\
\end{bmatrix}$$

2. Vector de coeficientes

$$\beta= \begin{bmatrix}
\beta_0\\
\beta_1\\
\vdots\\
\beta_n\\
\end{bmatrix}$$

3. Vector de errores

$$\epsilon= \begin{bmatrix}
\epsilon_0\\
\epsilon_1\\
\vdots\\
\epsilon_n\\
\end{bmatrix}$$

4. Vector de coeficientes dependientes

$$y= \begin{bmatrix}
y_0\\
y_1\\
\vdots\\
y_n\\
\end{bmatrix}$$

5. El método consiste en encontrar los coeficientes $\beta$ de tal manera que se minimicen los errores:

$$b=  \min_{\beta}(y-X^T\beta)^2$$

6. Los valores estimados se encuentran mediante la fórmula:

$$b = (X^TX)^{-1}X^TY$$
Este proceso nos entrega los coeficientes para los cuales se obtiene el error más bajo posible.

$$\hat{y}=b_0+b_1X_1+b_2X_2+\cdots+b_nX_n$$

La estimación de la varianza se obtiene como:

$$\sigma^2 =\cfrac{\sum (y_i-\hat{y_i})^2}{n-p}$$

Donde $n$ es la cantidad de observaciones y $p$ la cantidad de variables independientes.



### ¿Qué problemas pueden surgir?

1. **Multicolinealidad**

  Ocurre cuando dos o más variables independientes en el modelo están altamente correlacionadas, lo que puede hacer que la estimación de los coeficientes sea inestable y difícil de interpretar.

  Se detecta con la prueba  (*Factor de Inflación de la Varianza*). Un VIF>10 sugiere una fuerte multicolinealidad.

  Se soluciona eliminando variables, combinandolas o con técnicas de reducción de dimensionalidad como PCA.

2. **Autocorrelación**

  Presente cuando los residuos del modelo no son independientes entre sí, comúnmente en datos de series temporales.

  Se puede detectar con la prueba de *Durbin-Watson*. Valores del estadístico cerca de 2 indican ausencia de autocorrelación; valores significativamente menores o mayores indican autocorrelación positiva o negativa, respectivamente.

  Se soluciona usando técnicas de estimación robustas a la autocorrelación.

3. **Heteroscedasticidad**

   Ocurre cuando la varianza de los errores varía entre las observaciones, lo que puede llevar a estimaciones ineficientes de los coeficientes.

   Se detecta con una visualización de los residuos contra los valores ajustados o usando pruebas como *Breusch-Pagan* o *White*.

   Se soluciona haciendo transformaciones de las variables (logarítmica, raíz cuadrada), o usar métodos de estimación robustos como los Mínimos Cuadrados Ponderados.



## Ejemplo y desarrollo en Python

Vamos a simular un dataset que incluya las siguientes características:

* Número de observaciones: $n=1000$
* Variables independientes:
  * Edad: distribución normal con media 35 y desviación estándar 10.
  * Salario: distribución normal con media 50000 y desviación estándar 15000.
  * Años_de_Experiencia: dependiente linealmente de la edad más un término de error.
  * Horas_Trabajadas: relacionadas con los años de experiencia.

* Variable dependiente:
  Productividad: dependiente de todas las variables anteriores con una relación lineal más un término de error.

```python
import pandas as pd
import numpy as np

np.random.seed(42)  # Para reproducibilidad

# Número de observaciones
n = 1000

# Generar datos
edad = np.random.normal(35, 10, n)  # Edad
salario = np.random.normal(50000, 15000, n)  # Salario
años_de_experiencia = 0.4 * edad + np.random.normal(0, 2, n)  # Años de experiencia, correlacionados con la edad
horas_trabajadas = 40 + 0.5 * años_de_experiencia + np.random.normal(0, 3, n)  # Horas trabajadas

# Variable dependiente: Productividad
productividad = 5 + 0.03 * edad + 0.0004 * salario + 0.2 * años_de_experiencia + 0.05 * horas_trabajadas + np.random.normal(0, 1, n)

# Crear DataFrame
df = pd.DataFrame({
    'Edad': edad,
    'Salario': salario,
    'Años_de_Experiencia': años_de_experiencia,
    'Horas_Trabajadas': horas_trabajadas,
    'Productividad': productividad
})

# Mostrar las primeras filas del DataFrame
print(df.head())
```

## Matriz de correlaciones

```python
df.corr()
```

## Creación del modelo

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Ajustar el modelo de regresión lineal múltiple
X = df[['Edad', 'Salario', 'Años_de_Experiencia', 'Horas_Trabajadas']]  # Variables independientes
y = df['Productividad']  # Variable dependiente

# Añadir una constante al modelo (intercepto)
X = sm.add_constant(X)

# Crear un modelo OLS
modelo = sm.OLS(y, X)

# Ajustar el modelo
resultado = modelo.fit()

# Imprimir el resumen del modelo
print(resultado.summary())
```

### Predicciones

```python
# Nuevos datos para predicción
nuevos_datos = pd.DataFrame({
    'const': 1,  # Añadir el intercepto manualmente
    'Edad': [40],
    'Salario': [55000],
    'Años_de_Experiencia': [10],
    'Horas_Trabajadas': [45]
})

# Realizar predicciones
prediccion_productividad = resultado.predict(nuevos_datos)
print("La productividad predicha es:", prediccion_productividad[0])
```

```python
Y_pron = pd.DataFrame(resultado.predict(X_test), index=y_test.index, columns=["Pronostico"])
Y_pron["Reales"] = y_test
Y_pron["Media"] = np.mean(y_test)
Y_pron
```

1. **Multicolinealidad**

  Ocurre cuando dos o más variables independientes en el modelo están altamente correlacionadas, lo que puede hacer que la estimación de los coeficientes sea inestable y difícil de interpretar.

  Se detecta con la prueba  (*Factor de Inflación de la Varianza*). Un VIF>10 sugiere una fuerte multicolinealidad.

  Se soluciona eliminando variables, combinandolas o con técnicas de reducción de dimensionalidad como PCA.

3. **Heteroscedasticidad**

   Ocurre cuando la varianza de los errores varía entre las observaciones, lo que puede llevar a estimaciones ineficientes de los coeficientes.

   Se detecta con una visualización de los residuos contra los valores ajustados o usando pruebas como *Breusch-Pagan* o *White*.

```python
from statsmodels.stats.diagnostic import het_breuschpagan
# Realizar la prueba de Breusch-Pagan
bp_test = het_breuschpagan(residuos1, resultado1.model.exog)

# Mostrar los resultados de la prueba
bp_labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
print(dict(zip(bp_labels, bp_test)))
```
