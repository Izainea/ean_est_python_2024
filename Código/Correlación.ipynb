# Introducción a la Correlación

La correlación es una medida estadística que describe el grado de relación entre dos variables. Indica cómo el cambio en una variable está asociado con el cambio en otra. Es decir, muestra si los incrementos en una variable corresponden a incrementos en la otra (correlación positiva), si los incrementos en una variable corresponden a decrementos en la otra (correlación negativa), o si no hay una relación aparente (correlación cercana a cero).

Existen varios tipos de correlación, como la correlación de Pearson, la correlación de Spearman y la correlación de Kendall. pero vamos a iniciar mostrando la importancia de esta medida y algunos ejemplos sobre su uso.

### Importancia de la Correlación para la Estadística y la Analítica de Datos

La correlación es fundamental en estadística y análisis de datos por varias razones:

* **Descubrimiento de Relaciones**: Ayuda a identificar y cuantificar la fuerza de una relación entre dos variables, lo que es crucial para cualquier análisis exploratorio de datos.

* **Desarrollo de Modelos Predictivos**: La correlación es utilizada para seleccionar variables en modelos predictivos. Las variables que están fuertemente correlacionadas con la variable objetivo son a menudo candidatas para ser incluidas en el modelo.

* **Inferencia Causal**: Aunque la correlación por sí sola no implica causalidad, puede indicar posibles relaciones causales que merecen ser investigadas más a fondo.

* **Optimización y Mejora de Procesos**: En el ámbito empresarial y de la ingeniería, comprender las correlaciones entre diferentes variables puede ayudar a optimizar procesos y mejorar la eficiencia.

### Ejemplos de la Importancia de la Correlación

**Ejemplo 1**: Salud Pública


En el campo de la salud pública, la correlación se utiliza para estudiar la relación entre factores de riesgo y enfermedades. Por ejemplo, investigadores podrían encontrar una correlación positiva fuerte entre el consumo de tabaco y la incidencia de cáncer de pulmón en una población estudiada. Este tipo de análisis es crucial para desarrollar políticas de salud pública efectivas y para educar al público sobre comportamientos de riesgo.

**Ejemplo 2**: Marketing Digital


En marketing digital, la correlación se usa para entender las relaciones entre diversas actividades de marketing y sus impactos en las ventas o la conversión de clientes. Por ejemplo, un analista podría descubrir una correlación positiva entre el aumento del gasto en publicidad en redes sociales y el aumento en las ventas en línea. Este conocimiento permite a las empresas ajustar sus estrategias de marketing para maximizar el retorno sobre la inversión.

## Coeficiente de Correlación de Pearson

$$r = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^n (X_i - \bar{X})^2 \sum_{i=1}^n (Y_i - \bar{Y})^2}}$$

### ¿Cuándo se usa?

* Relación lineal entre variables.
* La distribución conjunta de las variables es normal.
* Homoscedasticidad, es decir la varianza de una variable es uniforme en todos los valores de la otra variable.
* Independencia de las Observaciones

El coeficiente de correlación de Pearson, también conocido como el coeficiente de correlación del producto-momento de Pearson, es una medida que cuantifica la relación lineal entre dos variables cuantitativas. Es uno de los indicadores más comunes y ampliamente utilizados para evaluar la fuerza y la dirección de una relación lineal entre dos variables.

### Rango de Valores y Significado

El coeficiente de Pearson puede tomar valores en el rango de $-1$ a $+1$, donde:

* $+1$ indica una correlación positiva perfecta: a medida que una variable aumenta, la otra también lo hace en una relación lineal perfecta.

* $-1$ indica una correlación negativa perfecta: a medida que una variable aumenta, la otra disminuye en una relación lineal perfecta.

* $0$ indica que no hay correlación lineal entre las dos variables.

Los valores intermedios indican el grado de correlación lineal entre las variables. Cuanto más cerca del -1 o +1, más fuerte es la correlación. Cuanto más cerca del 0, más débil es la correlación. Específicamente se dice que:

* Correlación Fuerte:
    * Positiva fuerte: $r$ está entre $+0.7$ y $+1.0$
    * Negativa fuerte: $r$ está entre $-0.7$ y $-1.0$

* Correlación Moderada:
    * Positiva moderada: $r$ está entre $+0.3$ y $+0.7$
    * Negativa moderada: $r$ está entre $-0.3$ y $-0.7$

* Correlación Débil:

    * Positiva débil: $r$ está entre $0$ y $+0.3$
    * Negativa débil: $r$ está entre $0$ y $-0.3$

* Sin Correlación: $r$ está muy cerca de $0$, indicando ninguna relación lineal.


### Ejemplos Contextuales

* En Psicología: Un $𝑟$ de +0.5 puede ser considerado una correlación fuerte debido a la variabilidad inherente al comportamiento humano y a los factores psicológicos.

* En Física: Las correlaciones pueden necesitar ser más cercanas a +1 o -1 para ser consideradas fuertes, debido a la precisión de las medidas y la expectativa de relaciones consistentes y replicables.

### Consideraciones Adicionales

* Significado Práctico: Además de la magnitud de $𝑟$, es importante considerar el tamaño de la muestra y la variabilidad de los datos. Un  $𝑟$ moderadamente alto en un estudio grande con datos consistentes puede ser más significativo que un $𝑟$ muy alto en un estudio pequeño con datos variables.

* Causalidad: Es crucial recordar que la correlación no implica causalidad, incluso si la correlación es muy fuerte. La relación observada puede ser influenciada por variables no medidas (variables confusas).

* Aplicabilidad: Los valores de correlación deben interpretarse dentro del contexto del área de estudio y del conocimiento existente sobre las variables involucradas. Lo que es considerado fuerte en una disciplina puede no serlo en otra.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos de ejemplo
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = 0.5 * x + np.random.normal(0, 1, 100)

# Calcular correlación
correlation = np.corrcoef(x, y)[0, 1]

# Crear scatter plot
plt.scatter(x, y, color='blue')
plt.title(f"Scatter Plot con Correlación de Pearson: r = {correlation:.2f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.show()
```

### Realizar una Prueba de Hipótesis para la Correlación de Pearson

La prueba de hipótesis para el coeficiente de correlación de Pearson es un procedimiento estadístico utilizado para determinar si la correlación observada entre dos variables en una muestra sugiere una correlación significativa entre esas variables en la población total, desde un punto de vista estadístico.

Pasos para Realizar una Prueba de Hipótesis para el Coeficiente de Pearson

1. Planteamiento de las hipótesis:

$$H_0 : r =0$$

$$H_1: r \neq 0$$

2. Calcular el Coeficiente de Correlación de Pearson:

$$r = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^n (X_i - \bar{X})^2 \sum_{i=1}^n (Y_i - \bar{Y})^2}}$$

3. Estadístico de prueba

El test estadístico para Pearson se puede calcular bajo la hipótesis nula de que no hay correlación en la población. Si 𝑛 es el tamaño de la muestra y
$r$  es el coeficiente de Pearson, el estadístico de prueba 𝑡 se calcula como:

$$t = \cfrac{r \sqrt{n-2}}{\sqrt{1-r^2}} = r \sqrt{\cfrac{n-2}{1-r^2}}$$



3. Determinar el $p-$valor.

Este estadístico sigue aproximadamente una distribución $t$ con $𝑛-2$ grados de libertad cuando $n$ es suficientemente grande y la hipótesis nula es cierta.

4. Compara el $p$ valor con el nivel de significancia establecido $\alpha$.



```python
import numpy as np
import scipy.stats as stats

# Datos de ejemplo
x = np.array([10, 20, 30, 40, 50])
y = np.array([12, 24, 33, 44, 55])

# Calcula el coeficiente de correlación de Pearson y el valor p asociado
pearson_corr, p_value = stats.pearsonr(x, y)

print(f"Pearson Correlation Coefficient: {pearson_corr:.3f}")
print(f"P-value: {p_value:.8f}")
```

Esto no indica que si existe una relación entre las variables y $r = 0.999$ nos indica que esta relación es casi perfecta.

## Coeficiente de Correlación de Spearman

El coeficiente de correlación de Spearman es una medida no paramétrica de la correlación entre dos variables ordinales o cuantitativas. A diferencia del coeficiente de Pearson, el coeficiente de Spearman no asume que las variables siguen una distribución normal y es más adecuado para evaluar la relación entre variables que no tienen una relación lineal.

### ¿Cuándo se usa?

El coeficiente de Spearman es particularmente útil en los siguientes casos:

* Cuando los datos no cumplen los supuestos necesarios para usar la correlación de Pearson, específicamente la normalidad de los datos.
* Con datos ordinales o cuando las variables no están en una escala de intervalo o de razón.
* Cuando se quiere evaluar relaciones no lineales entre dos variables, ya que Spearman es sensible a las relaciones monótonas.

### ¿Para qué se usa?

* Medir la Fuerza y Dirección de una Asociación Monótona: A diferencia de Pearson, que mide solo relaciones lineales, Spearman puede capturar relaciones monótonas, proporcionando una herramienta más flexible en ciertas situaciones.
* Análisis de Datos con Outliers: Spearman, al ser basado en rangos, es menos sensible a outliers en comparación con Pearson.
* Datos Categóricos Ordinales: Spearman es ideal para datos que se clasifican en categorías que tienen un orden natural (como rangos de satisfacción en encuestas).

### Fórmulación

$$\rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}$$

Donde,
* $d_i $ s la diferencia entre los rangos correspondientes de las dos variables para cada observación $𝑖$.
* $𝑛$ es el número total de observaciones.

### La fuerza de la correlación puede leerse en una tabla.

|Valor $\rho$	|Fuerza de la correlación|
|--------------|------------------------|
|0.0 - 0.1|	no hay correlación|
|0.1 - 0.3|	poca correlación|
|0.3 - 0.5|	correlación media|
|0.5 - 0.7|	correlación alta|
|0.7 - 1  |correlación muy alta|



### Asigmación de rangos y cálculo de $\rho$

1. Ordenar cada conjunto de datos y asignar rangos. En caso de empates, asigna a cada uno el promedio de los rangos que ocuparían.
2. Para cada observación, la diferencia $d_i = \text{Rango de } x_i-\text{Rango de } y_i$
​3. Obtener $d_i^2$
4. Calcular $\rho$

| A          | 82                     | 90                     |
| B          | 90                     | 85                     |
| C          | 76                     | 78                     |
| D          | 88                     | 91                     |

**Asignamos los rangos:**

| Estudiante | Rango X | Rango Y | $d_i$ | $d_i^2$ |
|------------|---------|---------|---------|-----------|
| A          | 2       | 4       | -2      | 4         |
| B          | 4       | 2.5     | 1.5     | 2.25      |
| C          | 1       | 1       | 0       | 0         |
| D          | 3       | 5       | -2      | 4         |

```python
n=5
d2=[4, 2.25, 0,  4, 0.25]
sum(d2)
```

```python
rho = 1-6*sum(d2)/(n*(n**2-1))
rho
```

El coeficiente de correlación de Spearman de  0.475 indica una correlación positiva moderada entre las puntuaciones en el examen X y el examen Y. Esto significa que existe una tendencia en la que estudiantes con puntuaciones más altas en un examen tienden a tener puntuaciones más altas en el otro, aunque esta relación no es fuertemente predictiva ni perfectamente monótona.

## Coeficiente de correlación de Kendall

El coeficiente de correlación de Kendall es una medida no paramétrica de la correlación entre dos variables ordinales o cuantitativas. Al igual que el coeficiente de Spearman, el coeficiente de Kendall no asume que las variables siguen una distribución normal y es más adecuado para evaluar la relación entre variables que no tienen una relación lineal.

### ¿Cuándo se usa?

El coeficiente de Kendall es particularmente útil en los siguientes casos:

* Cuando los datos no cumplen los supuestos necesarios para usar la correlación de Pearson, específicamente la normalidad de los datos.
* Con datos ordinales o cuando las variables no están en una escala de intervalo o de razón.
* Cuando se quiere evaluar relaciones no lineales entre dos variables, ya que Kendall es sensible a las relaciones monótonas.

### ¿Para qué se usa?

* Medir la Fuerza y Dirección de una Asociación Monótona: A diferencia de Pearson, que mide solo relaciones lineales, Kendall puede capturar relaciones monótonas, proporcionando una herramienta más flexible en ciertas situaciones.

* Análisis de Datos con Outliers: Kendall, al ser basado en rangos, es menos sensible a outliers en comparación con Pearson.

* Datos Categóricos Ordinales: Kendall es ideal para datos que se clasifican en categorías que tienen un orden natural (como rangos de satisfacción en encuestas).

### Fórmulación

$$\tau = \cfrac{C-D}{\cfrac{n(n-1)}{2}}$$

Donde,

* $C$ es el número de pares concordantes.
* $D$ es el número de pares discordantes.
* $n$ es el número total de observaciones.

### La fuerza de la correlación puede leerse en una tabla.

|Valor $\tau$	|Fuerza de la correlación|
|--------------|------------------------|
|0.0 - 0.1|	no hay correlación|
|0.1 - 0.3|	poca correlación|
|0.3 - 0.5|	correlación media|
|0.5 - 0.7|	correlación alta|
|0.7 - 1  |correlación muy alta|







### Asigmación de rangos y cálculo de $\tau$

1. Ordenar cada conjunto de datos y asignar rangos. En caso de empates, asigna a cada uno el promedio de los rangos que ocuparían.

2. Para cada par de observaciones, determina si son concordantes o discordantes. Un par es concordante si el orden relativo de las dos observaciones es el mismo en ambas variables, y discordante si el orden relativo es diferente.

3. Calcular $C$ y $D$.

4. Calcular $\tau$.

###  Ejemplo:
Supongamos que estamos interesados en estudiar la relación entre la antigüedad en el trabajo y la satisfacción laboral en una pequeña empresa.

Ambas variables serán tratadas como ordinales: la antigüedad se clasifica en categorías (1 = menos de un año, 2 = 1 a menos de 3 años, 3 = 3 a menos de 5 años, 4 = 5 años o más), y la satisfacción se mide con una escala de Likert (1 = muy insatisfecho, 2 = insatisfecho, 3 = neutral, 4 = satisfecho, 5 = muy satisfecho).





```python
Antigüedad: [3, 1, 4, 2, 3, 4, 2, 1]
Satisfacción: [4, 1, 5, 3, 4, 5, 2, 2]
```

```python
import scipy.stats as stats

# Datos de ejemplo
antiguedad = [3, 1, 4, 2, 3, 4, 2, 1]
satisfaccion = [4, 1, 5, 3, 4, 5, 2, 2]

# Calcular el coeficiente de correlación de Kendall y el valor p
tau, p_value = stats.kendalltau(antiguedad, satisfaccion)

print(f"Kendall's tau: {tau:.3f}")
print(f"P-value: {p_value:.4f}")
```

**Interpretación**

* Kendall's tau: Este valor indica la fuerza entre las dos variables es muy fuerte y en un sentido positivo, es decir, que entre más antiguedad más satisfacción hay entre los empleados.

* Valor-$p$: Proporciona información sobre la significancia estadística de la correlación observada. Igual que en los casos anteriores, este valor nos hace rechazar la hipótesis nula y no rechazar la idea de relación monótona entre las mismas.

```python
import numpy as np
import matplotlib.pyplot as plt
# Generar datos de ejemplo
np.random.seed(3)
x = np.array(antiguedad)
y = np.array(satisfaccion)

# Calcular correlación
tau, p_value = stats.kendalltau(antiguedad, satisfaccion)

# Crear scatter plot
plt.scatter(x, y, color='blue')
plt.title(f"Scatter Plot con Correlación de Kendall: tau = {tau:.3f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.show()
```

### Realizar una Prueba de Hipótesis para la Correlación de Kendall

Una prueba de hipótesis para la correlación de Kendall, estás probando si la correlación observada en tu muestra es significativamente diferente de cero (o de algún otro valor específico, aunque generalmente cero). El procedimiento típico implica los siguientes pasos:

1. Planteamiento de las hipótesis:

$$H_0 : \tau =0$$

$$H_1: \tau \neq 0$$

2. Calcular el Coeficiente de Correlación de Kendall:

$$\tau = \cfrac{C-D}{\cfrac{n(n-1)}{2}}$$

3. Estadístico de prueba

El test estadístico para Spearman se puede calcular bajo la hipótesis nula de que no hay correlación en la población. Si 𝑛 es el tamaño de la muestra y
𝜌  es el coeficiente de Spearman, el estadístico de prueba 𝑡 se calcula como:

$$Z = \frac{3 \tau \sqrt{n(n - 1)}}{\sqrt{2(2n + 5)}}$$


3. Determinar el $p-$valor.

El valor de Kendall's tau puede transformarse en un valor que siga aproximadamente una distribución normal bajo la hipótesis nula de que no hay correlación.

4. Compara el $p$ valor con el nivel de significancia establecido $\alpha$.

```python
import scipy.stats as stats
import numpy as np

# Datos de ejemplo
antiguedad = [3, 1, 4, 2, 3, 4, 2, 1]
satisfaccion = [4, 1, 5, 3, 4, 5, 2, 2]

# Calcular el coeficiente de correlación de Kendall y el valor p
tau, p_value = stats.kendalltau(antiguedad, satisfaccion)

# Número de observaciones
n = len(antiguedad)

# Calcular el estadístico Z
Z = (3 * tau * np.sqrt(n * (n - 1))) / np.sqrt(2 * (2 * n + 5))

print(f"Kendall's tau: {tau:.3f}")
print(f"Z-statistic: {Z:.3f}")
print(f"P-value: {p_value:.4f}")
```

Es decir, que si existe una relación positiva entre las variables y además esta relación es muy fuerte.

## Prueba de independencia

Una prueba de independencia entre dos variables es un método estadístico utilizado para determinar si **dos variables categóricas** son independientes una de la otra. La prueba más común para esto es la prueba de chi-cuadrado de independencia. Esta prueba examina si las distribuciones de frecuencia observadas difieren significativamente de las distribuciones de frecuencia que se esperarían si las dos variables fueran independientes.

### Prueba Chi-Cuadrado de Independencia

La prueba de chi-cuadrado de independencia se basa en la comparación de las frecuencias observadas y esperadas en una tabla de contingencia. La hipótesis nula es que las dos variables son independientes, mientras que la hipótesis alternativa es que hay una relación entre las dos variables.

### Pasos para Realizar una Prueba de Independencia

1. Planteamiento de las hipótesis:

$$H_0 : \text{Las dos variables son independientes}$$

$$H_1: \text{Las dos variables no son independientes}$$

2. Crear una tabla de contingencia:

Una tabla de contingencia es una tabla que muestra la distribución conjunta de dos variables categóricas. Las filas representan una variable y las columnas representan la otra variable. Cada celda de la tabla contiene el número de observaciones que caen en la intersección de una fila y una columna.

3. Calcular las frecuencias esperadas:

Las frecuencias esperadas son los valores que se esperarían si las dos variables fueran independientes. Se calculan multiplicando las frecuencias marginales de las filas y columnas correspondientes y dividiendo por el número total de observaciones.

4. Calcular el estadístico de prueba:

El estadístico de prueba para la prueba de chi-cuadrado de independencia se calcula como la suma de los cuadrados de las diferencias entre las frecuencias observadas y esperadas, dividida por las frecuencias esperadas.

$$\chi^2 = \sum \cfrac{(O_i - E_i)^2}{E_i}$$

Donde:

* $O_i$ es la frecuencia observada en la celda $i$.
* $E_i$ es la frecuencia esperada en la celda $i$.

5. Determinar el $p-$valor:

El estadístico de prueba sigue aproximadamente una distribución chi-cuadrado con $(r-1)(c-1)$ grados de libertad, donde $r$ es el número de filas y $c$ es el número de columnas en la tabla de contingencia.

6. Compara el $p$ valor con el nivel de significancia establecido $\alpha$.

## Pruebas de normalidad

Una prueba de normalidad es un procedimiento estadístico utilizado para determinar si una muestra de datos se ajusta a una distribución normal. Este tipo de pruebas es crucial en estadística porque muchas técnicas estadísticas asumen que los datos siguen una distribución normal. Si esta suposición no se cumple, los resultados de tales análisis pueden no ser válidos.

* Hipótesis Nula $(H_0)$: La muestra proviene de una distribución normal.

* Hipótesis Alternativa $(H_1)$: La muestra no proviene de una distribución normal.

### Métodos Comunes de Prueba de Normalidad

#####  Prueba de Shapiro-Wilk:

Esta es una de las pruebas más poderosas para la normalidad y es adecuada para *muestras pequeñas a moderadas*.

#####  Prueba de Kolmogorov-Smirnov:

Esta prueba compara la distribución empírica de los datos con una distribución normal teórica. Puede ser utilizada para *muestras más grande*s.

Incluye una opción para ajustar la media y la desviación estándar de la distribución normal a los datos.

#####  Prueba de Anderson-Darling:

Similar a la prueba de Kolmogorov-Smirnov pero con más peso en las colas de la distribución de los datos. Es más sensible a las desviaciones en las colas.

#####  QQ-Plot (Gráfico Cuantil-Cuantil):

Aunque técnicamente no es una prueba, es una herramienta visual poderosa para evaluar la normalidad. Comparan los cuantiles teóricos de una distribución normal con los cuantiles de los datos empíricos.

### Interpretación de los Resultados
* Rechazar la Hipótesis Nula $H_0$: Si el valor p de la prueba es menor que el nivel de significancia (comúnmente 0.05), rechazamos la hipótesis nula y concluimos que hay evidencia suficiente para afirmar que los datos no siguen una distribución normal.

* No Rechazar la Hipótesis Nula: Si el valor-$p$ es mayor que el nivel de significancia, no hay suficiente evidencia para rechazar la hipótesis nula, lo que sugiere que los datos pueden considerarse normalmente distribuidos.

#### Implementación:

```python
import scipy.stats as stats
import numpy as np

# Generar datos de ejemplo
data = np.random.normal(loc=0, scale=1, size=100)

# Realizar la prueba de Shapiro-Wilk
shapiro_stat, p_value = stats.shapiro(data)

print(f"Shapiro-Wilk Statistic: {shapiro_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
import scipy.stats as stats
import numpy as np

# Generar datos de ejemplo
data = np.random.uniform(0, 1, size=100)

# Realizar la prueba de Shapiro-Wilk
shapiro_stat, p_value = stats.shapiro(data)

print(f"Shapiro-Wilk Statistic: {shapiro_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
import scipy.stats as stats
import numpy as np

# Generar datos de ejemplo
data = np.random.poisson(1 ,size=100)

# Realizar la prueba de Shapiro-Wilk
shapiro_stat, p_value = stats.shapiro(data)

print(f"Shapiro-Wilk Statistic: {shapiro_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Continuamos usando los datos generados anteriormente

# Realizar la prueba de Kolmogorov-Smirnov (K-S) para una distribución normal
ks_stat, p_value = stats.kstest(data, 'norm', args=(np.mean(data), np.std(data)))
print(f"Kolmogorov-Smirnov Statistic: {ks_stat:.4f}, P-value: {p_value:.4f}")
```

```python
# Realizar la prueba de Anderson-Darling
ad_test = stats.anderson(data, dist='norm')
print(f"Anderson-Darling Statistic: {ad_test.statistic:.4f}")
for i in range(len(ad_test.critical_values)):
    sl, cv = ad_test.significance_level[i], ad_test.critical_values[i]
    print(f"At significance level {sl}%: CV = {cv:.4f}, {'Reject' if ad_test.statistic > cv else 'Accept'} the null hypothesis")
```

# Ejemplo

Vamos a trabajar con un conjunto de datos con las siguientes variables:

* Edad: Variable continua que indica la edad de los individuos.
* Ingresos: Variable continua con los ingresos de los individuos.
* Satisfacción Laboral: Variable ordinal que simula la satisfacción laboral en una escala de 1 a 5.
* Departamento: Variable categórica que simula diferentes departamentos en una empresa.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Configurar la semilla aleatoria para reproducibilidad
np.random.seed(42)

# Crear datos simulados
data = {
    'Edad': np.random.normal(35, 8, 100),  # Edad con media de 35 y desviación estándar de 8
    'Ingresos': np.random.normal(50000, 15000, 100),  # Ingresos en dólares
    'Satisfacción Laboral': np.random.choice([1, 2, 3, 4, 5], 100, p=[0.1, 0.2, 0.4, 0.2, 0.1]),  # Satisfacción Laboral
    'Departamento': np.random.choice(['Finanzas', 'Marketing', 'IT', 'RRHH'], 100)  # Departamentos
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Visualizar los primeros datos
print(df.head())
```

## Prueba de normalidad

```python
# Realizar prueba de normalidad en la variable Edad
print("Prueba de normalidad (Shapiro-Wilk) para Edad:")
print(stats.shapiro(df['Edad']))
```

```python
# Realizar un QQ-plot para visualizar la distribución de la edad
print("\nQQ-plot para Edad:")
stats.probplot(df['Edad'], dist="norm", plot=plt)
plt.title('QQ Plot - Edad')
plt.show()
```

```python
# Realizar prueba de normalidad en la variable Ingresos
print("Prueba de normalidad (Shapiro-Wilk) para Ingresos:")
print(stats.shapiro(df['Ingresos']))
```

```python
# Realizar un QQ-plot para visualizar la distribución de los Ingresos
print("\nQQ-plot para Ingresos:")
stats.probplot(df['Ingresos'], dist="norm", plot=plt)
plt.title('QQ Plot - Ingresos')
plt.show()
```

### ¿Será que las variables Edad e Ingresos tienen relación?

```python
import matplotlib.pyplot as plt
import numpy as np


x = df['Edad']
y = df['Ingresos']

# Calcular correlación
correlation = np.corrcoef(x, y)[0, 1]

# Crear scatter plot
plt.scatter(x, y, color='blue')
plt.title(f"Scatter Plot con Correlación de Pearson: r = {correlation:.2f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.show()
```

```python
spearman_corr, spearman_p_value = stats.spearmanr(df['Edad'], df['Ingresos'])
print(f"Spearman correlation coefficient: {spearman_corr:.3f}, P-value: {spearman_p_value:.3f}")
```

#### ¿Hay relación entre la satisfacción laboral y el departamento?

```python
# Realizar prueba de independencia entre Satisfacción Laboral y Departamento
ct = pd.crosstab(df['Satisfacción Laboral'], df['Departamento'])
ct
```

```python
chi2, p_value, dof, expected = stats.chi2_contingency(ct)
print("\nChi-squared test de independencia entre Satisfacción Laboral y Departamento:")
print(f"Chi-squared: {chi2:.2f}, P-value: {p_value:.3f}, Grados de libertad: {dof}")
```

```python
kendall_tau, kendall_p_value = stats.kendalltau(df['Satisfacción Laboral'], df['Departamento'])
print(f"Kendall's tau: {kendall_tau:.3f}, P-value: {kendall_p_value:.3f}")
```
