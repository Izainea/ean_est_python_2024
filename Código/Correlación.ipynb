## Coeficiente de Correlaci贸n de Pearson

$$r = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^n (X_i - \bar{X})^2 \sum_{i=1}^n (Y_i - \bar{Y})^2}}$$

### 驴Cu谩ndo se usa?

* Relaci贸n lineal entre variables.
* La distribuci贸n conjunta de las variables es normal.
* Homoscedasticidad, es decir la varianza de una variable es uniforme en todos los valores de la otra variable.
* Independencia de las Observaciones

El coeficiente de correlaci贸n de Pearson, tambi茅n conocido como el coeficiente de correlaci贸n del producto-momento de Pearson, es una medida que cuantifica la relaci贸n lineal entre dos variables cuantitativas. Es uno de los indicadores m谩s comunes y ampliamente utilizados para evaluar la fuerza y la direcci贸n de una relaci贸n lineal entre dos variables.

### Rango de Valores y Significado

El coeficiente de Pearson puede tomar valores en el rango de $-1$ a $+1$, donde:

* $+1$ indica una correlaci贸n positiva perfecta: a medida que una variable aumenta, la otra tambi茅n lo hace en una relaci贸n lineal perfecta.

* $-1$ indica una correlaci贸n negativa perfecta: a medida que una variable aumenta, la otra disminuye en una relaci贸n lineal perfecta.

* $0$ indica que no hay correlaci贸n lineal entre las dos variables.

Los valores intermedios indican el grado de correlaci贸n lineal entre las variables. Cuanto m谩s cerca del -1 o +1, m谩s fuerte es la correlaci贸n. Cuanto m谩s cerca del 0, m谩s d茅bil es la correlaci贸n. Espec铆ficamente se dice que:

* Correlaci贸n Fuerte:
    * Positiva fuerte: $r$ est谩 entre $+0.7$ y $+1.0$

* Correlaci贸n Moderada:
    * Positiva moderada: $r$ est谩 entre $+0.3$ y $+0.7$

* Correlaci贸n D茅bil:

    * Positiva d茅bil: $r$ est谩 entre $0$ y $+0.3$
    * Negativa d茅bil: $r$ est谩 entre $0$ y $-0.3$

* Sin Correlaci贸n: $r$ est谩 muy cerca de $0$, indicando ninguna relaci贸n lineal.


### Ejemplos Contextuales

* En Psicolog铆a: Un $$ de +0.5 puede ser considerado una correlaci贸n fuerte debido a la variabilidad inherente al comportamiento humano y a los factores psicol贸gicos.

* Significado Pr谩ctico: Adem谩s de la magnitud de $$, es importante considerar el tama帽o de la muestra y la variabilidad de los datos. Un  $$ moderadamente alto en un estudio grande con datos consistentes puede ser m谩s significativo que un $$ muy alto en un estudio peque帽o con datos variables.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos de ejemplo
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = 0.5 * x + np.random.normal(0, 1, 100)

# Calcular correlaci贸n
correlation = np.corrcoef(x, y)[0, 1]

# Crear scatter plot
plt.scatter(x, y, color='blue')
plt.title(f"Scatter Plot con Correlaci贸n de Pearson: r = {correlation:.2f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.show()
```

Pasos para Realizar una Prueba de Hip贸tesis para el Coeficiente de Pearson

2. Calcular el Coeficiente de Correlaci贸n de Pearson:

$$r = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum_{i=1}^n (X_i - \bar{X})^2 \sum_{i=1}^n (Y_i - \bar{Y})^2}}$$

$r$  es el coeficiente de Pearson, el estad铆stico de prueba  se calcula como:

$$t = \cfrac{r \sqrt{n-2}}{\sqrt{1-r^2}} = r \sqrt{\cfrac{n-2}{1-r^2}}$$



3. Determinar el $p-$valor.

Este estad铆stico sigue aproximadamente una distribuci贸n $t$ con $-2$ grados de libertad cuando $n$ es suficientemente grande y la hip贸tesis nula es cierta.

4. Compara el $p$ valor con el nivel de significancia establecido $\alpha$.

```python
import scipy.stats as stats
import numpy as np

# Datos de ejemplo
antiguedad = [3, 1, 4, 2, 3, 4, 2, 1]
satisfaccion = [4, 1, 5, 3, 4, 5, 2, 2]

# Calcular el coeficiente de correlaci贸n de Kendall y el valor p
tau, p_value = stats.kendalltau(antiguedad, satisfaccion)

# N煤mero de observaciones
n = len(antiguedad)

# Calcular el estad铆stico Z
Z = (3 * tau * np.sqrt(n * (n - 1))) / np.sqrt(2 * (2 * n + 5))

print(f"Kendall's tau: {tau:.3f}")
print(f"Z-statistic: {Z:.3f}")
print(f"P-value: {p_value:.4f}")
```

Es decir, que si existe una relaci贸n positiva entre las variables y adem谩s esta relaci贸n es muy fuerte.

## Prueba de independencia

Una prueba de independencia entre dos variables es un m茅todo estad铆stico utilizado para determinar si **dos variables categ贸ricas** son independientes una de la otra. La prueba m谩s com煤n para esto es la prueba de chi-cuadrado de independencia. Esta prueba examina si las distribuciones de frecuencia observadas difieren significativamente de las distribuciones de frecuencia que se esperar铆an si las dos variables fueran independientes.

### Prueba Chi-Cuadrado de Independencia

La prueba de chi-cuadrado de independencia se basa en la comparaci贸n de las frecuencias observadas y esperadas en una tabla de contingencia. La hip贸tesis nula es que las dos variables son independientes, mientras que la hip贸tesis alternativa es que hay una relaci贸n entre las dos variables.

### Pasos para Realizar una Prueba de Independencia

1. Planteamiento de las hip贸tesis:

$$H_0 : \text{Las dos variables son independientes}$$

$$H_1: \text{Las dos variables no son independientes}$$

2. Crear una tabla de contingencia:

Una tabla de contingencia es una tabla que muestra la distribuci贸n conjunta de dos variables categ贸ricas. Las filas representan una variable y las columnas representan la otra variable. Cada celda de la tabla contiene el n煤mero de observaciones que caen en la intersecci贸n de una fila y una columna.

3. Calcular las frecuencias esperadas:

Las frecuencias esperadas son los valores que se esperar铆an si las dos variables fueran independientes. Se calculan multiplicando las frecuencias marginales de las filas y columnas correspondientes y dividiendo por el n煤mero total de observaciones.

4. Calcular el estad铆stico de prueba:

El estad铆stico de prueba para la prueba de chi-cuadrado de independencia se calcula como la suma de los cuadrados de las diferencias entre las frecuencias observadas y esperadas, dividida por las frecuencias esperadas.

$$\chi^2 = \sum \cfrac{(O_i - E_i)^2}{E_i}$$

Donde:

* $O_i$ es la frecuencia observada en la celda $i$.
* $E_i$ es la frecuencia esperada en la celda $i$.

5. Determinar el $p-$valor:

El estad铆stico de prueba sigue aproximadamente una distribuci贸n chi-cuadrado con $(r-1)(c-1)$ grados de libertad, donde $r$ es el n煤mero de filas y $c$ es el n煤mero de columnas en la tabla de contingencia.

6. Compara el $p$ valor con el nivel de significancia establecido $\alpha$.

## Pruebas de normalidad

Una prueba de normalidad es un procedimiento estad铆stico utilizado para determinar si una muestra de datos se ajusta a una distribuci贸n normal. Este tipo de pruebas es crucial en estad铆stica porque muchas t茅cnicas estad铆sticas asumen que los datos siguen una distribuci贸n normal. Si esta suposici贸n no se cumple, los resultados de tales an谩lisis pueden no ser v谩lidos.

* Hip贸tesis Nula $(H_0)$: La muestra proviene de una distribuci贸n normal.

* Hip贸tesis Alternativa $(H_1)$: La muestra no proviene de una distribuci贸n normal.

### M茅todos Comunes de Prueba de Normalidad

#####  Prueba de Shapiro-Wilk:

Esta es una de las pruebas m谩s poderosas para la normalidad y es adecuada para *muestras peque帽as a moderadas*.

#####  Prueba de Kolmogorov-Smirnov:

Esta prueba compara la distribuci贸n emp铆rica de los datos con una distribuci贸n normal te贸rica. Puede ser utilizada para *muestras m谩s grande*s.

Incluye una opci贸n para ajustar la media y la desviaci贸n est谩ndar de la distribuci贸n normal a los datos.

#####  Prueba de Anderson-Darling:

Similar a la prueba de Kolmogorov-Smirnov pero con m谩s peso en las colas de la distribuci贸n de los datos. Es m谩s sensible a las desviaciones en las colas.

#####  QQ-Plot (Gr谩fico Cuantil-Cuantil):

Aunque t茅cnicamente no es una prueba, es una herramienta visual poderosa para evaluar la normalidad. Comparan los cuantiles te贸ricos de una distribuci贸n normal con los cuantiles de los datos emp铆ricos.

### Interpretaci贸n de los Resultados
* Rechazar la Hip贸tesis Nula $H_0$: Si el valor p de la prueba es menor que el nivel de significancia (com煤nmente 0.05), rechazamos la hip贸tesis nula y concluimos que hay evidencia suficiente para afirmar que los datos no siguen una distribuci贸n normal.

* No Rechazar la Hip贸tesis Nula: Si el valor-$p$ es mayor que el nivel de significancia, no hay suficiente evidencia para rechazar la hip贸tesis nula, lo que sugiere que los datos pueden considerarse normalmente distribuidos.

#### Implementaci贸n:

```python
import scipy.stats as stats
import numpy as np

# Generar datos de ejemplo
data = np.random.normal(loc=0, scale=1, size=100)

# Realizar la prueba de Shapiro-Wilk
shapiro_stat, p_value = stats.shapiro(data)

print(f"Shapiro-Wilk Statistic: {shapiro_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Continuamos usando los datos generados anteriormente

# Realizar la prueba de Kolmogorov-Smirnov (K-S) para una distribuci贸n normal
ks_stat, p_value = stats.kstest(data, 'norm', args=(np.mean(data), np.std(data)))
print(f"Kolmogorov-Smirnov Statistic: {ks_stat:.4f}, P-value: {p_value:.4f}")
```
