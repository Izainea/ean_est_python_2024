# Regresión

## ¿Qué es la Regresión?

La regresión es una técnica estadística y de aprendizaje automático utilizada para expresar la relación entre variables. Consiste en crear un modelo que explique cómo una variable dependiente cambia en función de una o más variables independientes. El objetivo principal es predecir valores futuros de la variable dependiente a partir de los valores conocidos de las variables independientes.

$$y = f(x)$$

$$z=f(x,y)$$

$$y=f(x_1, x_2, ..., x_n)$$



```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos los valores de x
x = np.linspace(-10, 10, 100)  # 100 puntos entre -10 y 10
y = 3 * x + 5

plt.figure(figsize=(8, 6))  # Tamaño de la gráfica
plt.plot(x, y, label='y = 3x + 5')  # Graficar la línea

# Añadimos título y etiquetas
plt.title('Gráfica de una Función Lineal')
plt.xlabel('x')
plt.ylabel('y')

# Añadimos una leyenda
plt.legend()

# Mostramos la gráfica
plt.grid(True)  # Añade una cuadrícula para facilitar la lectura
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creamos una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definimos el rango de x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Creamos la malla de puntos en el plano XY
X, Y = np.meshgrid(x, y)

# Calculamos los valores de Z basados en la ecuación del plano
Z = 3*X + 2*Y + 1

# Graficamos el plano
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='b')

# Añadimos etiquetas para los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Añadimos un título
ax.set_title('Gráfico de un Plano en 3D')

# Mostramos la gráfica
plt.show()
```

En la analítica de datos, la regresión se utiliza ampliamente para **predecir y pronosticar**. Por ejemplo, en el comercio minorista, una empresa podría usar regresión para prever las ventas futuras basándose en datos históricos de ventas, el clima, y otras variables económicas, permitiendo así una planificación de inventario más eficiente y una estrategia de precios más efectiva.

## Tipos de Regresión

Existen varios tipos de modelos de regresión, cada uno adaptado a diferentes tipos de datos y relaciones:

* **Regresión Lineal Simple y Múltiple**: Usada para modelar relaciones lineales entre dos o más variables.
* **Regresión Logística**: Utilizada cuando la variable dependiente es categórica, comúnmente para problemas de clasificación.
* **Regresión Polinomial**: Permite modelar relaciones no lineales a través de términos polinomiales.
* **Regresión Ridge y Lasso**: Técnicas que incorporan penalización en los coeficientes para reducir el sobreajuste y mejorar la generalización del modelo.
* **Regresión con Efectos Fijos y Efectos Aleatorios**: Especialmente útil en datos de panel, donde se puede considerar la variabilidad tanto entre individuos como a lo largo del tiempo.

Cada tipo de regresión tiene sus propios usos y ventajas, dependiendo de la naturaleza del problema y del tipo de datos disponibles. Esta diversidad en métodos de regresión permite a los analistas de datos seleccionar el enfoque más adecuado para cada situación específica, optimizando así la precisión y utilidad de sus modelos predictivos.


# Regresión lineal simple

La regresión lineal simple es una técnica estadística que se usa para estudiar la relación entre **dos variables cuantitativas**: una variable independiente (predictora) y una variable dependiente (respuesta).

En este modelo, se asume que la variable dependiente puede ser aproximada como una **función lineal** de la variable independiente. Esto significa que se busca entender cómo el cambio en la variable independiente afecta el valor esperado de la variable dependiente, suponiendo que esta relación es lineal.

### Conceptos relacionados

* $x$ variable independiente, que utilizamos como predictor.
* $y$ variable dependiente que queremos predecir.
* $S_{xy}$ covarianza entre $x$ y $y$.
* $r_{xy}$ el coeficiente de correlación de Pearson.


## Modelo de la regresión lineal simple​

La ecuación de un modelo de regresión lineal simple es de la forma:

$$y = \beta_0 +\beta_1 x + 𝜖$$

Donde:

* $\beta_0$  es la intercepción con el eje $y$ o el término constante. Representa el valor esperado de $y$ cuando $x=0$.

* $\beta_1$ es el coeficiente de la variable independiente. Representa la pendiente de la recta, es decir el cambio esperado en $y$ por una unidad de cambio en $x$.

* $ϵ$  es el término de error, que captura todas las otras influencias en $y$ que no son explicadas por $x$.

## Ecuación de regresión estimada (muestral)

La ecuación de regresión estimada (muestral) se obtiene a partir de los datos observados, y tiene la forma:

$$\hat{y} = b_0 + b_1 x$$

Donde:

* $b_0$ es la estimación de la intercepción $\beta_0$.

* $b_1$ es la estimación del coeficiente de la variable independiente $\beta_1$.

* $\hat{y}$ es el valor estimado de $y$ para un valor dado de $x$.

```python
# Generar datos de ejemplo
x = np.array([10, 20, 30, 40, 50])
y = np.array([12, 24, 33, 44, 55])

# Calcular correlación
correlation = np.corrcoef(x, y)[0, 1]

# Crear scatter plot
plt.scatter(x, y, color='blue')
plt.title(f"Scatter Plot con Correlación de Pearson: r = {correlation:.2f}")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.plot(x, 1.06*x+1.8, label='y = 1.06x + 1.8')  # Graficar la línea
plt.show()
```

## Errores en la regresión lineal

Los errores en la regresión lineal son las diferencias entre los valores observados y los valores predichos por el modelo. Estos errores se conocen como **residuos** y se utilizan para evaluar la precisión del modelo.

Los residuos se calculan restando los valores observados de los valores predichos:

$$e_i = y_i - \hat{y}_i$$

Donde:

* $e_i$ es el residuo para la observación $i$.

* $y_i$ es el valor observado de la variable dependiente para la observación $i$.

* $\hat{y}_i$ es el valor predicho de la variable dependiente para la observación $i$.

## Método de mínimos cuadrados

El método de mínimos cuadrados es una técnica utilizada para estimar los coeficientes de la regresión lineal. Consiste en encontrar los valores de los coeficientes que minimizan la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo.   

La ecuación de la recta de regresión estimada se obtiene minimizando la suma de los cuadrados de los residuos:

$$\sum_{i=1}^{n} e_i^2= \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Donde:

* $y_i$ es el valor observado de la variable dependiente para la observación $i$.

* $\hat{y}_i$ es el valor predicho de la variable dependiente para la observación $i$.

* $n$ es el número total de observaciones.

## Valores $b_0$ y $b:1$ de la regresión estimada

Los valores de $b_0$ y $b_1$ se obtienen a partir de las siguientes fórmulas:

$$b_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

$$b_0 = \bar{y} - b_1 \bar{x}$$

Observe como $b_1$ es la razón entre la covarianza $S_{xy}$ y la varianza de la variable independiente $S_x$. Mientras que $b_0$ sale de sustituir en la ecuación de la recta $y = b_0 +b_1x$ los valores promedio de $x$ y $y$.





$$y=b_1x+b_0$$

$$\bar{y}=b_1\bar{x}+b_0$$

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.array([10, 20, 30, 40, 50])
y = np.array([12, 24, 33, 44, 55])

# Añadir una constante a los datos de entrada para incluir el término de intercepción (beta_0) en el modelo
x = sm.add_constant(x)

# Ajustar el modelo de regresión lineal
model = sm.OLS(y, x) # Ordinary Least Squares, OLS (Mínimos Cuadrados Ordinarios)
results = model.fit() # Ajustar el modelo a los datos

# Imprimir los resultados del modelo
print(results.summary().tables[1])

# Gráfico de los datos y la línea de ajuste
plt.scatter(np.array([10, 20, 30, 40, 50]), y, color='blue')  # datos originales
plt.plot(np.array([10, 20, 30, 40, 50]), results.predict(x), color='red')  # línea de ajuste
plt.title('Regresión Lineal Simple')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Errores en la regresión lineal

En el caso de la regresión lineal existen tres tipos de errores:

* Suma de cuadrados debida al error:
$$ SCE = \sum (y_i-\hat{y_i})^2$$

* Suma total de cuadrados:

$$ SCT = \sum (y_i-\bar{y})^2$$

* Suma de cuadrados debida al modelo o debida a la regresión:

$$ SCR = \sum (\hat{y_i}-\bar{y})^2$$

## Coeficiente de determinación $R^2$

El coeficiente de determinación $R^2$ es una medida de la proporción de la variabilidad en la variable dependiente que es explicada por el modelo de regresión.

El coeficiente de determinación $R^2$ se calcula como:

$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

$$R^2 = \frac{\sum_{i=1}^{n} (y_i - \bar{y})^2-\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

$$R^2 = \cfrac{SCR}{STC}$$

Donde:

* $y_i$ es el valor observado de la variable dependiente para la observación $i$.

* $\hat{y}_i$ es el valor predicho de la variable dependiente para la observación $i$.

* $\bar{y}$ es la media de los valores observados de la variable dependiente.

* $n$ es el número total de observaciones.

El coeficiente de determinación $R^2$ varía entre 0 y 1, y cuanto más cercano a 1, mejor es el ajuste del modelo a los datos.











## Coeficiente de correlación

Una forma análoga de encontrar el coeficiente de correlación de Pearson es:

$$r_{xy}  = (\text{Signo de } b_1)\sqrt{\text{Coeficiente de determinación}}$$

$$r_{xy}  = (\text{Signo de } b_1)\sqrt{R^2}$$

## Distribución muestral de $\beta_1$

Tenga en cuenta, que estimamos el parámeto $\beta_1$ por medio de la pendiente de la recta $b_1$. Nos interesa, estudiar la distribución muestral de $\beta_1$ y los valores que puede tomar $b_1$ para aproximar a $\beta_1$ con cierto grado de confianza.

Tenemos que listribución muestral de $\beta_1$ es una **distribución normal** con:

* Valor esperado:
$$E(\beta_1) = \beta_1$$
* Desviación estándar

$$\sigma_{\beta_1}= \cfrac{\sigma}{\sqrt{\sum{(x_i-\bar{x})^2}}}$$

* Desviación estándar estimada

$$s_{b_1}= \cfrac{s}{\sqrt{\sum{(x_i-\bar{x})^2}}}$$

* Error cuadrado medio

El error cuadrático medio (ECM) es una medida de la calidad de un estimador. Se calcula como la media de los errores al cuadrado, es decir, la diferencia entre los valores observados y los valores predichos por el modelo.

El error cuadrático medio (ECM) se calcula como:

$$s^2 =\cfrac{SCE}{n-2} = \cfrac{\sum (y_i-\hat{y_i})^2}{n-2}$$

* Error estandar de la estimación

$$s= \sqrt{ECM}$$

Este valor, es la desviación estándar de los errores. Es una medida de la precisión de las predicciones del modelo. Cuanto menor sea el error estándar de la estimación, mejor será la precisión del modelo.

## Prueba de significacia

En el caso de la regresión lineal, nos interesa saber si la relación entre las variables es significativa. Para ello, se realiza una prueba de hipótesis para determinar si el coeficiente de la variable independiente es significativamente diferente de cero.

La hipótesis nula es que el coeficiente de la variable independiente es igual a cero, lo que significa que no hay relación entre las variables. La hipótesis alternativa es que el coeficiente es diferente de cero, lo que indica que hay una relación significativa entre las variables.

$$H_0: \beta_1 = 0$$

$$H_1: \beta_1 \neq 0$$

Es decir es una prueba a dos colas.

Para realizar la prueba de significancia, se calcula el estadístico $t$ y se compara con el valor crítico de la distribución $t$ con $n-2$ grados de libertad.

$$t = \frac{b_1}{s_{b_1}}$$

Donde:

* $b_1$ es el coeficiente de la variable independiente.

* $s_{b_1}$ es el error estándar del coeficiente de la variable independiente.

* $n$ es el número total de observaciones.

Si el valor absoluto del estadístico $t$ es mayor que el valor crítico de la distribución $t$, se rechaza la hipótesis nula y se concluye que el coeficiente de la variable independiente es significativamente diferente de cero.

## Intervalo de confianza

El intervalo de confianza es un rango de valores dentro del cual se espera que esté el verdadero valor del coeficiente de la variable independiente con cierto nivel de confianza.

El intervalo de confianza se calcula como:

$$b_1 \pm t_{\alpha/2} s_{b_1}$$

Donde se usa la distribución $t$ no $n-2$ grados de libertad.

En nuestro caso el intervalo de confianza es:

$$(0.987, 1.133)$$

Por defecto el paquete trabaja con una confianza del $95\%$

Tenemos un conjunto de datos simple que consiste en el tamaño de diferentes casas (en pies cuadrados) y sus precios correspondientes (en miles de dólares).


```python
x = np.array([600, 800, 1000, 1200, 1400, 1600, 1800, 2000])
y = np.array([150, 180, 210, 240, 270, 300, 330, 360])
correlation = np.corrcoef(x, y)[0, 1]
correlation
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos de ejemplo
# Tamaños de la casa en pies cuadrados (como variable independiente)
X = np.array([600, 800, 1000, 1200, 1400, 1600, 1800, 2000]).reshape(-1, 1)

# Precios de la casa en miles de dólares (como variable dependiente)
Y = np.array([150, 180, 210, 240, 270, 300, 330, 360])

# Crear el modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo a los datos
model.fit(X, Y)

# Predecir los precios usando el modelo ajustado
Y_pred = model.predict(X)

# Coeficientes del modelo
print("Intercepto:", model.intercept_)
print("Pendiente:", model.coef_[0])

# Graficar los datos y la línea de ajuste
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, Y_pred, color='red', label='Línea de ajuste')
plt.title('Regresión Lineal de Precios de Casas vs. Tamaño')
plt.xlabel('Tamaño de la Casa (pies cuadrados)')
plt.ylabel('Precio de la Casa (miles de dólares)')
plt.legend()
plt.show()
```
